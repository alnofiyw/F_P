```python
# السؤال الثاني عن تحليل البيانات 
# استخدم هذه البيانات وقم باستيرادها 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel('Adidas DataSet.xlsx')
df
```

```python
# 1. unique valves for' Region' and their count
df['Region'].unique()
```

```python
df['Region'].value_counts()
```

```python
# 2. Pivot table for Sum of 'Units Sold' per 'State'
df.pivot_table(index='State', values='Units Sold', aggfunc='sum').sort_values(by='Units Sold', ascending=False)
```

```python
df
```

```python
# 3. Bar Chart for Sum of 'Total Sales' per 'Product'

sales_product = df.groupby('Product')['Total Sales'].sum()
sales_product
```

```python
plt.figure(figsize=(10, 6))
plt.bar(sales_product.index, sales_product.values)
plt.title('Total Sales per Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Price per Unit'].mean()
```

```python
# 4.Filter the data based on 'Price per Unit' greater than the mean of 'Price per Unit'
df[df['Price per Unit'] > df['Price per Unit'].mean()]
```

```python
df
```

```python
# 5. Do a linear regression analysis between "Price per Unit' and 'Operating Profit'
plt.figure(figsize=(8, 6))
sns.regplot(x='Price per Unit', y='Operating Profit', data=df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.xlabel('Price per Unit')
plt.ylabel('Operating Profit')
plt.title('Linear Regression: Price per Unit vs Operating Profit')
plt.grid(True)
plt.show()
```
